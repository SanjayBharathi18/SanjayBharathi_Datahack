# Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression

# To ignore the warnings that result from undertaking certain operations.
import warnings
warnings.filterwarnings("ignore")
```

# Importing Data

```python
# Training DataSets
train_feature = pd.read_csv(r"D:\GFG\training_set_features.csv")
train_target=pd.read_csv(r"D:\GFG\training_set_labels.csv")

# Testing DataSet
df_test = pd.read_csv(r"D:\GFG\test_set_features.csv")

# Submission Format
submission=pd.read_csv(r"D:\GFG\submission_format.csv")
```

## Merging the Training Datasets

```python
df = train_feature.merge(train_target,on="respondent_id")
```

## Assigning New Index
- "respondent_id" is the new index because it has the unique id and we need to avoid this feature while training a model 

```python
df.set_index("respondent_id",inplace=True)
df.head()
```

## Checking for Null Values

```python
df.isnull().sum()
```

- We are going to fill the null values by taking mode() value for each feature

```python
categories=df.columns
categories
```

```python
for x in categories:
    cat_mode = df[x].mode()[0]
    df[x].fillna(cat_mode, inplace=True)
```

```python
df.isnull().sum()
```

- There is no null values

## Checking for Duplicate Values

```python
df.duplicated().any()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(keep=False,inplace=True)
```

```python
df.duplicated().any()
```

- We have removed the duplicate Values

```python
df.shape
```

## OneHot Encoding
- We have some categorical features which has values in text format
- ML models do not process these text format data (categorical data)
- So, we have to convert them into numerical data
- Therefore, for Nominal data we use OneHot Encoding method.

```python
categorical_df=df.select_dtypes(include="object")
categorical=categorical_df.columns
categorical
```

```python
for x in categorical:
    one_hot_encoder = OneHotEncoder(sparse=False)
    one_hot_encoded = one_hot_encoder.fit_transform(df[[x]])

    # Convert the result to a DataFrame and concatenate with the original DataFrame
    one_hot_encoded_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out([x]))
    one_hot_encoded_df.index=df.index
    df = pd.concat([df, one_hot_encoded_df], axis=1)
    df.drop(x, axis=1, inplace=True)
```

- We have converted all the categorical data to numerical data
- The Preprocessing of training data was done

## Checking for Null Values

```python
df_test.isnull().sum()
```

- We are going to fill the null values by taking mode() value for each feature

```python
categories=df_test.columns
categories
```

```python
for x in categories:
    cat_mode = df_test[x].mode()[0]
    df_test[x].fillna(cat_mode, inplace=True)
```

## OneHot Encoding

```python
categorical_df = df_test.select_dtypes(include="object")
categorical = categorical_df.columns
categorical
```

```python
for x in categorical:
    one_hot_encoder = OneHotEncoder(sparse=False)
    one_hot_encoded = one_hot_encoder.fit_transform(df_test[[x]])

    # Convert the result to a DataFrame and concatenate with the original DataFrame
    one_hot_encoded_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out([x]))
    one_hot_encoded_df.index = df_test.index
    df_test = pd.concat([df_test, one_hot_encoded_df], axis=1)
    df_test.drop(x, axis=1, inplace=True)
```

## Building ML Model

```python
X_train = df.drop(["xyz_vaccine","seasonal_vaccine"],axis=1)
Y_train_xyz = df[["xyz_vaccine"]]
Y_train_seas = df[["seasonal_vaccine"]]
```

## Model - xyz vaccine

```python
model_xyz=LogisticRegression(random_state=42,max_iter=10000)
model_xyz.fit(X_train,Y_train_xyz)
```

## Model - seasonal vaccine

```python
model_seas=LogisticRegression(random_state=42,max_iter=10000)
model_seas.fit(X_train,Y_train_seas)
```

```python
predict_xyz=model_xyz.predict_proba(df_test)[:,1]
predict_xyz
```

```python
predict_seas=model_seas.predict_proba(df_test)[:,1]
predict_seas
```

## Adding predicted values to the submission dataframe

```python
submissions = pd.DataFrame({
    'respondent_id': submission['respondent_id'],
    'xyz_vaccine': predict_xyz,
    'seasonal_vaccine': predict_seas
})
```
